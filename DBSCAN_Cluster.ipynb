```python
#fas-정재환
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import numpy as np
import csv
import json
import seaborn as sns
import mglearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
%matplotlib inline 
```

```python
#한글 폰트 적용 - 한글깨짐 방지
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv("NHIS_OPEN_GJ_2017_3.csv", encoding='cp949', nrows=7000)
df
```

```python
x = df[['신장', '체중', '허리둘레', '식전혈당']] #4개의 특성을 가진 데이터
scaler = StandardScaler() # 분산을1, 평균0으로 변환하는 스케일러
scaler.fit(x)
x_scaled = scaler.transform(x) # 변환
```

```python
dbscan = DBSCAN() 
clusters = dbscan.fit_predict(x_scaled) # DBSCAN 클러스터 예측
```

```python
# DBSCAN 예측 결과
plt.scatter(x_scaled[:,0], x_scaled[:,1], c=clusters, cmap=mglearn.cm2, s=100, edgecolors='black')
plt.xlabel('특성 0')
plt.ylabel('특성 1')
```
